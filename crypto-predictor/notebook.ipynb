```python
import pandas as pd
xrp = pd.read_csv("data/xrp_usd.csv")
btc = pd.read_csv("data/btc_usd.csv")
```

```python
def normalize(df):
    df = df.drop(["Vol.", "Price", "Open", "High", "Low"], axis=1)
    df = df.rename(index=str, columns={"Change %": "Change"})
    # "-2.60" becomes a float instead of a string
    df["Change"] = df["Change"].apply(lambda string: string[:-1]).values.astype(float)
    df["Date"] = pd.to_datetime(df["Date"])
    return df
```

```python
xrp = normalize(xrp)
btc = normalize(btc)
```
